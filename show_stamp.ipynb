```python
import os
import numpy as np
from PIL import Image
from astropy.io import fits
import matplotlib.pyplot as plt
```

```python
source_name = "J115252.26+004733.24"
```

```python
#show fits stamp 
now_path = os.getcwd()
path_fits = now_path+"/fits/"+source_name+".fits"
hud = fits.open(path_fits)
img = hud[0].data
img_scaled = [abs(intensity)**0.3 for intensity in img]
plt.imshow(img_scaled, cmap='gray')
plt.colorbar()
plt.show()
```

```python
#show RGB stamp
path_rgb = now_path+"/RGB/"+source_name+".tif"
image = Image.open(path_rgb)
plt.xticks([])
plt.yticks([])
plt.imshow(np.asarray(image))
```
